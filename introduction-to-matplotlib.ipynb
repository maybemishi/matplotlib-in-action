##Matplotlib Example Workflow

```python
#0. import matplotlib and get it ready for plotting in jupyter

%matplotlib inline
import matplotlib.pyplot as plt

#1. get data ready
x=[1,2,3,4]
y=[11,22,33,44]

#2. setup plot (we use object oriented method)
fig,ax= plt.subplots(figsize=(10,10)) #width then height

#3. plot data
ax.plot(x,y);

#4. customise plot
ax.set( title= "simple plot",
       xlabel= "x-axis",
       ylabel="y-axis");

#5. save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")

```

##plotting from pandas dataframe

```python
import pandas as pd

#make a datframe
car_sales=pd.read_csv("car-sales.csv")
car_sales

```

```python
ts= pd.Series (np.random.randn(1000),index=pd.date_range("1/1/2025",periods=1000))
ts= ts.cumsum()
ts.plot();
```

```python
#car_sales['Price']= car_sales['Price'].str.replace('[\$\,\.]','')
car_sales['Price'] = car_sales['Price'].str.replace(r'[\$,\.]', '', regex=True)

car_sales
```

```python
#remove last two zeroes
car_sales['Price']= car_sales['Price'].str[:-2]
car_sales
```

```python
car_sales["sale date"]= pd.date_range("1/1/2020", periods= len(car_sales))
car_sales
```

```python
car_sales["total"]=car_sales['Price'].cumsum(0)
car_sales #price still in string format
```

```python
car_sales["total"]=car_sales['Price'].astype(int).cumsum(0)#coverted to integer
car_sales
```

```python
#let us plot the total sales
car_sales.plot(x="sale date", y="total");
```

```python
car_sales["Odometer (KM)"].plot.hist();
```

```python
car_sales["Odometer (KM)"].plot.hist(bins=50);
```

```python
# plotting from pandas dataframe 4
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
#create a histogram of age
heart_disease["age"].plot.hist(bins=70); #histogram shows frequency distribution
```

```python
#subplots from dataframe
heart_disease.plot.hist(figsize=(10,100),subplots= True);
```

```python
over_50= heart_disease[heart_disease["age"]>50]
#oo method mixed with pyplot
fix, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax
            );

#ax.set_xlim([45,100]) #sets limit from 45 to 100
```

#oo method from scratch


```python
fig, (ax0,ax1)= plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10))

scatter= ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                    c=over_50["target"])

ax0.set(title="heart disease and cholestrol levels",
       xlabel="age",
       ylabel="cholestrol");
ax0.axhline(over_50["chol"].mean(),
            linestyle="--");
#ax0.legend(*scatter.legend_elements(),title="target");


scatter_2= ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                    c=over_50["target"])

ax1.set(title="heart disease and cholestrol levels",
       xlabel="age",
       ylabel="thalach");
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--")

ax1.legend(*scatter.legend_elements(),title="target");

#adding title to figure

fig.suptitle("Heart Disease Analysis",
             fontsize=16,
             weight="bold");
```
